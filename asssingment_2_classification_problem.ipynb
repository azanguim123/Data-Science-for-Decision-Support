```python
# Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC


from google.colab import files
#uploaded = files.upload()

from google.colab import drive
drive.mount('/content/drive')
#Load the dataset
df = pd.read_csv('/content/HR_comma_sep.csv')


csv_file_path = '/content/drive/My Drive/HR_comma_sep.csv'
df.to_csv(csv_file_path, index=False)

# Data Preprocessing
print("Dataset shape:", df.shape)
print("\nDataset preview:")
print(df.head())

# Remove unnecessary columns
if 'id' in df.columns:
    df.drop('id', axis=1, inplace=True)

# Check for missing values:
print("\nNull value counts:")
print(df.isnull().sum())

# Encode categorical variables
categorical_cols = df.select_dtypes(include='object').columns
if len(categorical_cols) > 0:
    print("\nCategorical columns found:", categorical_cols)
    df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Define features (X) and target (Y)
X = df.drop('left', axis=1)
Y = df['left']

# Output the number of classes
print("\nNumber of classes in target variable:", Y.nunique())

# Split data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Models for classification
models = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Support Vector Machine': SVC(random_state=42)
}

# Train and evaluate models
results = []
for model_name, model in models.items():
    model.fit(X_train_scaled, Y_train)
    train_accuracy = accuracy_score(Y_train, model.predict(X_train_scaled))
    test_accuracy = accuracy_score(Y_test, model.predict(X_test_scaled))
    results.append({
        'Model': model_name,
        'Train Accuracy': train_accuracy,
        'Test Accuracy': test_accuracy
    })
    # Confusion Matrix
    print(f"\n{model_name} Confusion Matrix:")
    cm = confusion_matrix(Y_test, model.predict(X_test_scaled))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix for {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Results summary
results_df = pd.DataFrame(results)
print("\nModel Evaluation Results:")
print(results_df)

# Hyperparameter optimization
optimized_results = []
for model_name, model in models.items():
    param_grid = {}
    if model_name == 'Logistic Regression':
        param_grid = {'C': [0.1, 1, 10]}
    elif model_name == 'Decision Tree':
        param_grid = {'max_depth': [5, 10, 15]}
    elif model_name == 'K-Nearest Neighbors':
        param_grid = {'n_neighbors': [3, 5, 7]}
    elif model_name == 'Support Vector Machine':
        param_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}

    grid = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
    grid.fit(X_train_scaled, Y_train)

    best_model = grid.best_estimator_
    train_accuracy = accuracy_score(Y_train, best_model.predict(X_train_scaled))
    test_accuracy = accuracy_score(Y_test, best_model.predict(X_test_scaled))
    optimized_results.append({
        'Model': model_name,
        'Best Params': grid.best_params_,
        'Train Accuracy': train_accuracy,
        'Test Accuracy': test_accuracy
    })

# Optimized results summary
optimized_results_df = pd.DataFrame(optimized_results)
print("\nOptimized Model Evaluation Results:")
print(optimized_results_df)

# K-Fold Cross Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
kfold_results = []
for model_name, model in models.items():
    fold_accuracies = []
    for train_index, test_index in kf.split(X):
        X_kf_train, X_kf_test = X.iloc[train_index], X.iloc[test_index]
        Y_kf_train, Y_kf_test = Y.iloc[train_index], Y.iloc[test_index]
        model.fit(X_kf_train, Y_kf_train)
        accuracy = accuracy_score(Y_kf_test, model.predict(X_kf_test))
        fold_accuracies.append(accuracy)
        # Confusion Matrix for each fold
        cm = confusion_matrix(Y_kf_test, model.predict(X_kf_test))
        print(f"\nConfusion Matrix for {model_name} (Fold):")
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title(f'Fold Confusion Matrix for {model_name}')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
    kfold_results.append({
        'Model': model_name,
        'Mean Accuracy': np.mean(fold_accuracies)
    })

# Display K-Fold results
kfold_results_df = pd.DataFrame(kfold_results)
print("\nK-Fold Cross-Validation Results:")
print(kfold_results_df)

# Analyze wrongly classified observations
wrongly_classified = {}  # Dictionary to store indices of wrongly classified observations

for model_name, model in models.items():
    predictions = model.predict(X_test_scaled)
    misclassified_indices = np.where(predictions != Y_test)[0]  # Indices where predictions are incorrect
    wrongly_classified[model_name] = misclassified_indices
    print(f"\n{model_name}: {len(misclassified_indices)} wrongly classified observations.")

# Find observations misclassified by all models
common_errors = set(wrongly_classified['Logistic Regression'])
for model_name in wrongly_classified.keys():
    common_errors = common_errors.intersection(set(wrongly_classified[model_name]))

print(f"\nNumber of observations wrongly classified by all models: {len(common_errors)}")
if len(common_errors) > 0:
    print("Indices of these observations:", common_errors)
else:
    print("No observations were misclassified by all models.")

```
